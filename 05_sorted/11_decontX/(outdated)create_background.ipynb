```python
# load foreground and raw matrix
import scanpy as sc

fdata = sc.read_h5ad("data/foreground.h5ad")
fdata.var_names_make_unique()
rdata = sc.read_10x_h5("data/raw.h5")
rdata.var_names_make_unique()

print("foreground shape: ", fdata.shape)
print("rawdata shape: ", rdata.shape)

# subtract foreground from raw
bdata = rdata.concatenate(fdata, join='outer')

print("background shape: ", bdata.shape)

# save background
#bdata.write_h5ad("data/background.h5ad")
```

```python
fcell_names = fdata.obs_names.tolist()
print(len(fcell_names))
```

```python
rcell_names = rdata.obs_names.tolist()
print(len(rcell_names))
```

```python
print("wir sollten also bei ", len(rcell_names)-len(fcell_names), " vielen Zellen landen")
```

```python
# check if all fcell_names are in rcell_names
count = 1
for fcell in fcell_names:
    if fcell not in rcell_names:
        print("fcell_name ", fcell, " wasnt found in raw")
    elif count == len(fcell_names):
        print("all fcell_names were found in raw")
    else:
        count +=    1
    
print("finished")
```

```python
# subtract all obs_varnames from rdata which are in fcell_names
bdata = rdata[~rdata.obs_names.isin(fcell_names), :]
print(bdata.shape)
# save rdata
bdata.write_h5ad("data/background.h5ad")
```
